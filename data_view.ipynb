```python
import pandas as pd
import numpy as np
import json
from collections import Counter
import MeCab
```

```python
with open("data/compound_train.json", 'r') as f:
    train_raw = json.load(f)['entry']
```

```python
flatten = lambda multi_list: [item for sublist in multi_list for item in sublist if (not isinstance(item, str)) or (len(item) is not 0)]
```

```python
uses = flatten([entry['Attributes']['用途'] for entry in train_raw])
```

```python
m = MeCab.Tagger("-Ochasen")
```

```python
def parse(text):
    node = m.parseToNode(text)
    words = []
    while node:
        if len(node.surface) == 0:
            node = node.next
            continue

        hinshi = node.feature.split(',')
        if hinshi[0] == '名詞':
            words.append(node.surface)

        node = node.next

    return words
```

```python
wakati_uses = flatten([parse(use) for use in uses])
```

```python
Counter(wakati_uses).most_common()
```
